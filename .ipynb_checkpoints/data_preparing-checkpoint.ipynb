```python
from datasets import load_dataset
import numpy as np
import matplotlib.pyplot as plt
```

```python
imdb_data_train = load_dataset("stanfordnlp/imdb")['train']
```

```python
sentiments = np.array(imdb_data_train['label']).squeeze()
plt.figure(figsize=(8, 6))
plt.hist(sentiments, bins=[0, 0.5, 1], rwidth=0.8, align='mid')
plt.xticks([0, 1])
plt.xlabel('Sentiment')
plt.ylabel('Frequency')
plt.title('Histogram of Sentiments')

# Display the plot
plt.show()
```
